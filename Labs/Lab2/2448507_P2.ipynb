```python
#import the neccessary libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the video
video_path = "sekiro.mp4"
cap = cv2.VideoCapture(video_path)
```

```python
#Check if video opened successfully
if not cap.isOpened():
    print("Error: Cannot open video.")
    exit()

frame_interval = 60  # Extract a frame every 30 frames
frame_count = 0
extracted_frames = []


#go to a loop to read all the frames, when there are no frames ret will become False and break.
#otherwise append the frame for every 60 frames
while True:
    ret, frame = cap.read()
    if not ret:
        break  # Exit loop if no more frames
    
    if frame_count % frame_interval == 0:
        extracted_frames.append(frame)
    
    frame_count += 1
#ensuring the file is properly closed
cap.release()
```

```python
# Process and display extracted frames
for i, frame in enumerate(extracted_frames):
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    red_channel = frame_rgb.copy()
    red_channel[:, :, 1:] = 0  # Keep only the red channel
    
    green_channel = frame_rgb.copy()
    green_channel[:, :, [0, 2]] = 0  # Keep only the green channel
    
    blue_channel = frame_rgb.copy()
    blue_channel[:, :, :2] = 0  # Keep only the blue channel
    
    # Display the original frame and color channels
    fig, axes = plt.subplots(1, 4, figsize=(12, 4))
    axes[0].imshow(frame_rgb)
    axes[0].set_title("Original Frame")
    axes[1].imshow(red_channel)
    axes[1].set_title("Red Channel")
    axes[2].imshow(green_channel)
    axes[2].set_title("Green Channel")
    axes[3].imshow(blue_channel)
    axes[3].set_title("Blue Channel")
    
    for ax in axes:
        ax.axis("off")
    
    plt.show()
```
