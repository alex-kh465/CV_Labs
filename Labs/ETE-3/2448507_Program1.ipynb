```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load traffic sign image
image = cv2.imread("program1.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# Apply Built-in Sobel Operator
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

# User-defined Sobel Operator using convolution
sobel_x_kernel = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_y_kernel = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

sobel_x_custom = cv2.filter2D(image, -1, sobel_x_kernel)
sobel_y_custom = cv2.filter2D(image, -1, sobel_y_kernel)
sobel_custom_combined = np.sqrt(sobel_x_custom**2 + sobel_y_custom**2)

# Display Results
fig, axes = plt.subplots(2, 3, figsize=(12, 6))

titles = [
    "Original", "Sobel X (Built-in)", "Sobel Y (Built-in)", 
    "Sobel X (Custom)", "Sobel Y (Custom)", "Sobel Combined"
]
images = [image, sobel_x, sobel_y, sobel_x_custom, sobel_y_custom, sobel_custom_combined]

# Loop through subplots and display images
for ax, img, title in zip(axes.ravel(), images, titles):
    ax.imshow(img, cmap='gray')
    ax.set_title(title, fontsize=12, pad=10)  # Increase padding
    ax.axis("off")  # Hide axes for better clarity

# Adjust layout to prevent overlap
plt.tight_layout()
plt.subplots_adjust(hspace=0.3, wspace=0.3)  # Increase spacing between plots

plt.show()
```

```python

# Built-in Prewitt Operator (Using Filter2D)
prewitt_x_kernel = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewitt_y_kernel = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])

prewitt_x = cv2.filter2D(image, -1, prewitt_x_kernel)
prewitt_y = cv2.filter2D(image, -1, prewitt_y_kernel)
prewitt_combined = np.sqrt(prewitt_x**2 + prewitt_y**2)
```

```python

# Compare with Sobel
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = np.sqrt(sobel_x**2 + sobel_y**2)

# Display
plt.figure(figsize=(10,5))
plt.subplot(2,3,1), plt.imshow(image, cmap='gray'), plt.title("Original X-ray")
plt.subplot(2,3,2), plt.imshow(prewitt_x, cmap='gray'), plt.title("Prewitt X")
plt.subplot(2,3,3), plt.imshow(prewitt_y, cmap='gray'), plt.title("Prewitt Y")
plt.subplot(2,3,4), plt.imshow(sobel_combined, cmap='gray'), plt.title("Sobel Edge Detection")
plt.subplot(2,3,5), plt.imshow(prewitt_combined, cmap='gray'), plt.title("Prewitt Edge Detection")
plt.show()
```

```python
# Built-in Roberts Operator
roberts_x_kernel = np.array([[1, 0], [0, -1]])
roberts_y_kernel = np.array([[0, 1], [-1, 0]])

roberts_x = cv2.filter2D(image, -1, roberts_x_kernel)
roberts_y = cv2.filter2D(image, -1, roberts_y_kernel)
roberts_combined = np.sqrt(roberts_x**2 + roberts_y**2)

# Display
plt.figure(figsize=(10,4))
plt.subplot(1,3,1), plt.imshow(image, cmap='gray'), plt.title("Original")
plt.subplot(1,3,2), plt.imshow(roberts_combined, cmap='gray'), plt.title("Roberts Edge Detection")
plt.show()
```

```python

# Apply Laplacian
laplacian = cv2.Laplacian(image, cv2.CV_64F, ksize=31)

# Display
plt.figure(figsize=(6,4))
plt.subplot(1,2,1), plt.imshow(image, cmap='gray'), plt.title("Original")
plt.subplot(1,2,2), plt.imshow(laplacian, cmap='gray'), plt.title("Laplacian Edge Detection")
plt.show()
```

```python
# Apply Canny Edge Detection
canny_edges = cv2.Canny(image, threshold1=50, threshold2=150)

# Display
plt.figure(figsize=(6,4))
plt.subplot(1,2,1), plt.imshow(image, cmap='gray'), plt.title("Original MRI")
plt.subplot(1,2,2), plt.imshow(canny_edges, cmap='gray'), plt.title("Canny Edge Detection")
plt.show()
```
