```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('Program-1.jpg')  # Replace with your image path
h, w = image.shape[:2]  # Get original dimensions

```

```python

# Define new size (double the original size)
new_size = (2 * w, 2 * h)

# Apply interpolation techniques
nearest = cv2.resize(image, new_size, interpolation=cv2.INTER_NEAREST)
bilinear = cv2.resize(image, new_size, interpolation=cv2.INTER_LINEAR)
bicubic = cv2.resize(image, new_size, interpolation=cv2.INTER_CUBIC)
lanczos = cv2.resize(image, new_size, interpolation=cv2.INTER_LANCZOS4)
```

```python
import matplotlib.pyplot as plt

# nearest, bilinear, bicubic, and lanczos
images = [nearest, bilinear, bicubic, lanczos]
titles = ['Nearest Neighbor', 'Bilinear', 'Bicubic', 'Lanczos']

# Plot the images using plt
plt.figure(figsize=(12, 6))

for i in range(4):
    plt.subplot(1, 4, i + 1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct display
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

Analysis of Interpolation Techniques
5. Which interpolation technique results in the sharpest image but might introduce artifacts like ringing?

Lanczos Interpolation

    It provides the sharpest image but can introduce ringing artifacts (halo effect around edges).
    Best for high-quality applications where sharpness is a priority.

6. Which technique produces the smoothest transitions but might blur fine details?

Bilinear Interpolation

    It results in a smooth image with less noticeable artifacts, but fine details may blur.

7. If the image contains text, which method best preserves the edges of the letters?

Nearest Neighbor Interpolation

    Best for text and pixelated images because it preserves hard edges, avoiding blurring.

8. For real-time applications where speed is a priority, which method is the most efficient?

Nearest Neighbor Interpolation

    Fastest because it only considers the nearest pixel without complex calculations.

9. If the goal is to create a high-quality print, which interpolation method is recommended and why?

Bicubic Interpolation or Lanczos Interpolation

    Bicubic: Balances sharpness and smooth transitions, reducing jagged edges.
    Lanczos: Best for preserving details in high-quality printing, but might introduce artifacts.
